```python
import pandas as pd
import numpy as np

credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')

movies = movies.merge(credits,on='title')
print(movies.head(1))
```

```python
#genre,movie_id, keywords, title, overview, cast, crew
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(1)
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def convertToList(obj,name=''):
  lst = []
  if isinstance(obj, list):
    return obj
  if name == 'cast':
    obj = ast.literal_eval(obj)[:3]
    for c in obj:
      lst.append(c['character'])
  elif name == 'crew':
    for crew in ast.literal_eval(obj):
      if crew['job'].lower() == 'director':
        lst.append(crew['name'])
        break
  else:
    for g in ast.literal_eval(obj):
      lst.append(g['name'])
    # return [i for i in ast.literal_eval(genres)]
  return lst
```

```python
movies['genres'] = movies['genres'].apply(convertToList)
```

```python
movies['keywords'] = movies['keywords'].apply(convertToList)
```

```python
movies['cast'] = movies.cast.apply(convertToList,args = ('cast',))
```

```python
movies['crew'] = movies['crew'].apply(convertToList, args = ('crew',))
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head(3)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
movies_new_df = movies[['movie_id', 'title', 'tags']]
```

```python
movies_new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
  y = []
  for i in obj.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
movies_new_df['tags'] = movies_new_df['tags'].apply(lambda x: ' '.join(x).lower())
movies_new_df['tags']
```

```python
movies_new_df['tags'] = movies_new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movies_new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()[:200]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
  movie_index = movies_new_df[movies_new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]

  for i in movie_list:
    print(movies_new_df.iloc[i[0]]['title'])
```

```python
recommend('Batman')
```
